```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, Dense, LayerNormalization
from tensorflow.keras.models import Model
import numpy as np

# Define Transformer model
def build_transformer(input_dim, output_dim, seq_length):
    inputs = Input(shape=(seq_length,))
    embedding_layer = Embedding(input_dim=input_dim, output_dim=output_dim)(inputs)
    transformer_block = tf.keras.layers.MultiHeadAttention(num_heads=8, key_dim=64)(embedding_layer, embedding_layer)
    transformer_block = LayerNormalization()(transformer_block)
    output = Dense(1, activation='sigmoid')(transformer_block)
    
    model = Model(inputs=inputs, outputs=output)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Example dataset
X_train = np.random.randint(0, 5000, size=(1000, 100))
y_train = np.random.randint(0, 2, size=(1000, 1))

# Initialize and train the model
transformer_model = build_transformer(input_dim=5000, output_dim=64, seq_length=100)
transformer_model.fit(X_train, y_train, epochs=10, batch_size=32)

# Summary of the model
transformer_model.summary()
```
