```python
{
 "cells": [
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "# 🖼️ CNN for Image Classification\n",
    "\n",
    "Train a Convolutional Neural Network (CNN) on the CIFAR-10 dataset to classify images.\n",
    "\n",
    "### 📚 What You'll Learn:\n",
    "- Load and preprocess the CIFAR-10 dataset.\n",
    "- Build a CNN model using TensorFlow/Keras.\n",
    "- Train the CNN model on the training data.\n",
    "- Evaluate the model's performance on the test data."
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 1: Import necessary libraries\n",
    "import tensorflow as tf\n",
    "from tensorflow.keras.datasets import cifar10\n",
    "from tensorflow.keras.models import Sequential\n",
    "from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense\n",
    "from tensorflow.keras.utils import to_categorical"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 2: Load and preprocess the data\n",
    "# Load the CIFAR-10 dataset\n",
    "(x_train, y_train), (x_test, y_test) = cifar10.load_data()\n",
    "\n",
    "# Normalize pixel values to the range [0, 1]\n",
    "x_train, x_test = x_train / 255.0, x_test / 255.0\n",
    "\n",
    "# Convert labels to one-hot encoded vectors\n",
    "y_train, y_test = to_categorical(y_train), to_categorical(y_test)"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 3: Build the CNN model\n",
    "model = Sequential([\n",
    "    # Convolutional layer with 32 filters, 3x3 kernel, and ReLU activation\n",
    "    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),\n",
    "    # Max pooling layer with 2x2 pool size\n",
    "    MaxPooling2D((2, 2)),\n",
    "    # Convolutional layer with 64 filters, 3x3 kernel, and ReLU activation\n",
    "    Conv2D(64, (3, 3), activation='relu'),\n",
    "    # Max pooling layer with 2x2 pool size\n",
    "    MaxPooling2D((2, 2)),\n",
    "    # Flatten the feature maps\n",
    "    Flatten(),\n",
    "    # Dense layer with 64 units and ReLU activation\n",
    "    Dense(64, activation='relu'),\n",
    "    # Output layer with 10 units (one for each class) and softmax activation\n",
    "    Dense(10, activation='softmax')\n",
    "])\n",
    "\n",
    "# Compile the model\n",
    "model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 4: Train the model\n",
    "# Train the model for 10 epochs, using 20% of the data for validation\n",
    "model.fit(x_train, y_train, epochs=10, validation_split=0.2)"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 5: Evaluate the model\n",
    "# Evaluate the model on the test data\n",
    "loss, acc = model.evaluate(x_test, y_test)\n",
    "print(f\"Test accuracy: {acc:.2f}\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### ✅ Key Takeaways:\n",
    "- CNNs are powerful models for image classification tasks.\n",
    "- The CIFAR-10 dataset is a common benchmark for image classification.\n",
    "- Experiment with different architectures, hyperparameters, and datasets to improve performance.\n",
    "\n",
    "Feel free to explore more advanced CNN architectures like ResNet or Inception for better results!"
   ]
  }
 ],
 "metadata": {
  "kernelspec": {
   "display_name": "Python 3",
   "language": "python",
   "name": "python3"
  },
  "language_info": {
   "name": "python",
   "version": ""
  }
 },
 "nbformat": 4,
 "nbformat_minor": 5
}
```
