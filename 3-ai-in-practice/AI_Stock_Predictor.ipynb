```python
# AI_Stock_Predictor.ipynb

# 📊 Predicting Stock Prices using Linear Regression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Fake stock data
days = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
prices = np.array([100, 102, 104, 103, 108])

model = LinearRegression()
model.fit(days, prices)

future_day = [[6]]
predicted_price = model.predict(future_day)

print(f"📉 Predicted price on Day 6: ${predicted_price[0]:.2f}")

# Plot
plt.plot(days, prices, marker='o', label='Actual Prices')
plt.plot(future_day, predicted_price, 'rx', label='Predicted Day 6')
plt.xlabel('Day')
plt.ylabel('Price')
plt.legend()
plt.title('Simple Stock Predictor')
plt.show()
```
