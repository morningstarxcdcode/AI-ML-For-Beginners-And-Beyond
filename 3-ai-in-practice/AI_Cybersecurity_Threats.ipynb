```python
# AI_Cybersecurity_Threats.ipynb

# 🔐 AI in Cybersecurity: Threat Detection with Machine Learning
# Objective: Train a model to detect suspicious network activity.

import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Sample dataset (mocked for demo)
data = {
    'duration': [0, 5, 10, 0, 15],
    'protocol': [1, 2, 1, 1, 2],
    'src_bytes': [100, 5000, 200, 0, 1234],
    'dst_bytes': [0, 1000, 20, 5000, 234],
    'threat': [0, 1, 0, 1, 0]  # 0=Normal, 1=Threat
}

df = pd.DataFrame(data)

X = df.drop('threat', axis=1)
y = df['threat']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("🔎 Classification Report:\n")
print(classification_report(y_test, y_pred))
```
