```python
{
 "cells": [
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "# 🔐 AI in Cybersecurity: Threat Detection with Machine Learning\n",
    "\n",
    "Train a model to detect suspicious network activity using machine learning.\n",
    "\n",
    "### 🎯 Objective:\n",
    "- Build a Random Forest Classifier to identify network threats.\n",
    "- Evaluate the model's performance using a classification report."
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 1: Import necessary libraries\n",
    "import pandas as pd\n",
    "from sklearn.ensemble import RandomForestClassifier\n",
    "from sklearn.model_selection import train_test_split\n",
    "from sklearn.metrics import classification_report"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 2: Create a sample dataset (mocked for demo)\n",
    "# This dataset simulates network traffic data with features like duration, protocol, and bytes transferred\n",
    "data = {\n",
    "    'duration': [0, 5, 10, 0, 15],\n",
    "    'protocol': [1, 2, 1, 1, 2],\n",
    "    'src_bytes': [100, 5000, 200, 0, 1234],\n",
    "    'dst_bytes': [0, 1000, 20, 5000, 234],\n",
    "    'threat': [0, 1, 0, 1, 0]  # 0=Normal, 1=Threat\n",
    "}\n",
    "\n",
    "# Create a Pandas DataFrame from the sample data\n",
    "df = pd.DataFrame(data)"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 3: Prepare the data for training\n",
    "# Separate features (X) from the target variable (y)\n",
    "X = df.drop('threat', axis=1)  # Features: duration, protocol, src_bytes, dst_bytes\n",
    "y = df['threat']  # Target variable: threat (0 or 1)\n",
    "\n",
    "# Split the dataset into training and testing sets\n",
    "X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)  # 70% training, 30% testing"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 4: Train the Random Forest Classifier\n",
    "# Create a Random Forest Classifier model\n",
    "model = RandomForestClassifier(random_state=42)  # Use a random state for reproducibility\n",
    "\n",
    "# Train the model on the training data\n",
    "model.fit(X_train, y_train)"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# Step 5: Evaluate the model\n",
    "# Make predictions on the test data\n",
    "y_pred = model.predict(X_test)\n",
    "\n",
    "# Print the classification report to evaluate the model's performance\n",
    "print(\"\\n🔎 Classification Report:\\n\")\n",
    "print(classification_report(y_test, y_pred))"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### ✅ Key Takeaways:\n",
    "- Machine learning can be used to detect network threats.\n",
    "- The Random Forest Classifier is a powerful algorithm for classification tasks.\n",
    "- The classification report provides insights into the model's precision, recall, and F1-score.\n",
    "\n",
    "Feel free to experiment with different datasets, features, or algorithms to improve the model's performance!"
   ]
  }
 ],
 "metadata": {
  "kernelspec": {
   "display_name": "Python 3",
   "language": "python",
   "name": "python3"
  },
  "language_info": {
   "name": "python",
   "version": ""
  }
 },
 "nbformat": 4,
 "nbformat_minor": 5
}
```
